```python
for col in category_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
    df_test[col] = df_test[col].fillna(df_test[col].mode()[0])
```

```python
def target_encode_train_test(train, test, cat_cols, target, n_splits=5, random_state=42):
    import numpy as np
    from sklearn.model_selection import KFold

    train_encoded = train.copy()
    test_encoded = test.copy()
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    for col in cat_cols:
        # KFold Ð´Ð»Ñ train
        train_encoded[col + "_te"] = 0
        for train_idx, valid_idx in kf.split(train):
            fold_train = train.iloc[train_idx]
            fold_valid = train.iloc[valid_idx]
            means = fold_train.groupby(col)[target].mean()
            train_encoded.loc[valid_idx, col + "_te"] = fold_valid[col].map(means)
        # unseen ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸Ð¸
        train_encoded[col + "_te"].fillna(train[target].mean(), inplace=True)

        # Ð´Ð»Ñ test Ð¿Ñ€Ð¾ÑÑ‚Ð¾ mapping Ð¿Ð¾ train
        means_all = train.groupby(col)[target].mean()
        test_encoded[col + "_te"] = test[col].map(means_all)
        test_encoded[col + "_te"].fillna(train[target].mean(), inplace=True)

    return train_encoded, test_encoded
```

```python
df_encoded, df_test_encoded = target_encode_train_test(
    df, df_test, category_columns, 'SalePrice'
)
```

```python
df_encoded.drop(columns=category_columns, inplace=True)
df_test_encoded.drop(columns=category_columns, inplace=True)
```

```python
X = df_encoded.drop(columns=['Id', 'SalePrice'], axis = 1)
y = df_encoded['SalePrice']
```

```python
import lightgbm as lgb

model = lgb.LGBMRegressor()
model.fit(X, y)
importances = pd.Series(model.feature_importances_, index=X.columns)
print(importances.sort_values(ascending=False))
```

```python
threshold = 14
important_features = importances[importances > threshold].index
X_selected = X[important_features]
X_selected
```

```python
id_part = df_test_encoded['Id']
X_test = df_test_encoded.drop(columns = ['Id'], axis =1)
X_selected = X_test[important_features]
```

```python
import xgboost as xgb
import optuna
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

def objective(trial):
    param = {
        'verbosity': 0,
        'objective': 'reg:squarederror',
        'tree_method': 'hist',  # Ð±Ñ‹ÑÑ‚Ñ€Ð¾Ðµ Ð¿Ð¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð´ÐµÑ€ÐµÐ²ÑŒÐµÐ²
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 5),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 5),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 5)
    }

    model = xgb.XGBRegressor(**param, random_state=42)

    # KFold CV
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_selected, y, cv=kf, scoring='neg_root_mean_squared_error')
    return -np.mean(scores)  # Optuna Ð¼Ð¸Ð½Ð¸Ð¼Ð¸Ð·Ð¸Ñ€ÑƒÐµÑ‚
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)  # ÐœÐ¾Ð¶Ð½Ð¾ ÑƒÐ²ÐµÐ»Ð¸Ñ‡Ð¸Ñ‚ÑŒ n_trials Ð´Ð»Ñ Ð»ÑƒÑ‡ÑˆÐµÐ³Ð¾ Ð¿Ð¾Ð¸ÑÐºÐ°

print("Ð›ÑƒÑ‡ÑˆÐ¸Ðµ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹:", study.best_params)
print("Ð›ÑƒÑ‡ÑˆÐ¸Ð¹ RMSE:", study.best_value)
```

```python
best_params = study.best_params
final_model = xgb.XGBRegressor(**best_params, random_state=42)
final_model.fit(X, y)

X_test = df_test_encoded.drop('SalePrice', axis=1, errors='ignore')
preds = final_model.predict(X_test.drop('Id', axis = 1))
```

```python
submission = pd.DataFrame({
    'Id': X_test['Id'],
    'SalePrice': preds
})
submission.to_csv('sub.csv', index = False)
```
