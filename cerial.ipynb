```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['type'])
plt.title('Density Distribution of type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes([int, float]).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='mfr', y='sugars', data=df)
plt.title('Распределение сахара по производителям')
plt.xlabel('Производитель')
plt.ylabel('Содержание сахара (грамм)')
plt.show()
```

```python
df_melt = df.melt(id_vars='mfr', value_vars=['sugars','protein','calories'],
                  var_name='Nutrient', value_name='Value')

plt.figure(figsize=(12,6))
sns.boxplot(x='mfr', y='Value', hue='Nutrient', data=df_melt)
plt.title('Распределение питательных веществ по производителям')
plt.show()
```

```python
df.drop('HealthSum', axis = 1, inplace = True)
```

```python
df.head()
```

```python
X = df.drop(['name','mfr','type'], axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Clusters'] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Clusters'])
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PC1','PC2'])
pca_df['mfr'] = df['mfr']
plt.figure(figsize=(10,6))
sns.scatterplot(x='PC1', y='PC2', hue='mfr', data=pca_df, palette='Set2')
plt.title('PCA визуализация злаков по питательным характеристикам')
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

tsne_df = pd.DataFrame(X_tsne, columns=['Dim1','Dim2'])
tsne_df['mfr'] = df['mfr']

plt.figure(figsize=(10,6))
sns.scatterplot(x='Dim1', y='Dim2', hue='mfr', data=tsne_df, palette='Set2')
plt.title('t-SNE визуализация злаков по питательным характеристикам')
plt.show()
```
